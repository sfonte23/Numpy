#Numpy
NumPy (Numerical Python) é uma biblioteca fundamental para computação científica em Python. É amplamente utilizada para manipulação de arrays e matrizes multidimensionais, além de fornecer uma grande coleção de funções matemáticas para operar nesses arrays.

##**Exemplo Prático:**

'Maria é uma analista de dados em uma empresa de tecnologia que desenvolve soluções de previsão do tempo. Recentemente, a empresa firmou uma parceria com uma rede de escolas públicas para fornecer previsões climáticas diárias, ajudando a planejar atividades ao ar livre e garantir a segurança dos alunos.'

**`O Problema:`**
Durante uma reunião com a direção da escola, foi identificado que as variações extremas de temperatura durante a primavera estavam dificultando o planejamento de eventos escolares. As temperaturas variavam significativamente de um dia para o outro, e as previsões precisavam ser mais precisas e acessíveis.

Maria recebeu uma semana de dados de temperaturas diárias em Celsius e foi incumbida de analisar esses dados para entender melhor os padrões climáticos e oferecer previsões mais precisas.

**`A Solução:`**
Maria decidiu usar a biblioteca NumPy e MatplotLib para realizar a análise dos dados de temperatura e fornecer insights valiosos.

```python
import numpy as np

# Carregar dados de um arquivo CSV usando NumPy
dados = np.genfromtxt('temperaturas.csv', delimiter=';', skip_header=1, dtype=None, encoding='utf-8')

# Verificar se os dados foram carregados corretamente
temperaturas_celsius = (dados[:90])
print(temperaturas_celsius)
```

Cálculo Médio das Temperaturas

```python
# Extrair as temperaturas da matriz de tuplas
temperaturas_celsius = np.array([linha[1] for linha in dados])

# Calcular a média das temperaturas
media_temperaturas = np.mean(temperaturas_celsius)

# Imprimir a média das temperaturas
print(f"Média das temperaturas: {media_temperaturas:.2f}°C")
```

Cálculo de Variância das Temperaturas

```python
variacao_temperaturas = np.var(temperaturas_celsius)
print(f"Variação das temperaturas: {variacao_temperaturas:.2f}")
```

Dias mais quentes e frios (Amplitude)

```python
# Identificar o índice do dia mais quente
indice_dia_mais_quente = np.argmax(temperaturas_celsius)
dia_mais_quente = indice_dia_mais_quente + 1  # Adicionamos 1 porque os dias começam do índice 0

# Identificar o índice do dia mais frio
indice_dia_mais_frio = np.argmin(temperaturas_celsius)
dia_mais_frio = indice_dia_mais_frio + 1  # Adicionamos 1 porque os dias começam do índice 0

print(f"Dia mais quente: Dia {dia_mais_quente} - Temperatura: {temperaturas_celsius[indice_dia_mais_quente]}°C")
print(f"Dia mais frio: Dia {dia_mais_frio} - Temperatura: {temperaturas_celsius[indice_dia_mais_frio]}°C")
```

**Visualização dos Dados**
> Para facilitar a compreensão dos dados pela equipe da escola, Maria criou alguns gráficos.



```python
import matplotlib.pyplot as plt

# Gráfico das temperaturas em Celsius
plt.figure(figsize=(10, 5))
plt.plot(temperaturas_celsius, label='Temperatura em °C',linestyle='--')
plt.xlabel('Dia')
plt.ylabel('Temperatura (°C)')
plt.title('Temperaturas Diárias em °C')
plt.legend()
plt.show()
```

Entendendo que a temperatura ideal está entre 22 e 27 ºC, vamos inserir na visualização

```python
# Plotar um gráfico das temperaturas ao longo do próximo ano
plt.figure(figsize=(12, 6))
plt.plot(temperaturas_celsius, color='blue')
plt.axhline(y=22, color='green', linestyle='--', label='Limite Inferior (22°C)')
plt.axhline(y=27, color='orange', linestyle='--', label='Limite Superior (27°C)')
plt.xlabel('Dia do Ano')
plt.ylabel('Temperatura (°C)')
plt.title('Temperaturas ao Longo do Próximo Ano')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Identificar os dias com temperaturas ideais (22-27°C)
indices_dias_ideais = np.where((temperaturas_celsius >= 22) & (temperaturas_celsius <= 27))[0]
dias_ideais = indices_dias_ideais + 1  # Adicionamos 1 porque os dias começam do índice 0

print("Melhores dias para programar atividades ao ar livre:")
for dia in dias_ideais:
    print(f"Dia {dia}")
```

```python
import matplotlib.pyplot as plt

# Suponha que você tenha os dados das temperaturas e dos dias ideais

# Plotar os pontos correspondentes aos dias ideais
plt.figure(figsize=(10, 6))
plt.scatter(dias_ideais, temperaturas_celsius[dias_ideais - 1], color='blue', label='Dias Ideais')

# Adicionar linhas verticais para separar os intervalos
plt.axvline(x=31, color='red', linestyle='--', linewidth=1)
plt.axvline(x=61, color='red', linestyle='--', linewidth=1)

plt.xlabel('Dia do Ano')
plt.ylabel('Temperatura (°C)')
plt.title('Dias Ideais para Atividades ao Ar Livre')
plt.legend()
plt.show()
```

```python
# Contar o número de dias ideais em cada intervalo
dias_1_30 = sum(1 for dia in dias_ideais if dia <= 30)
dias_31_60 = sum(1 for dia in dias_ideais if 31 <= dia <= 60)
dias_61_90 = sum(1 for dia in dias_ideais if 61 <= dia <= 90)

# Determinar o intervalo com mais dias ideais
if dias_1_30 > dias_31_60 and dias_1_30 > dias_61_90:
    print("O intervalo 1-30 tem mais dias ideais.Total de", dias_1_30, " dias")
elif dias_31_60 > dias_1_30 and dias_31_60 > dias_61_90:
    print("O intervalo 31-60 tem mais dias ideais.Total de", dias_31_60, " dias")
else:
    print("O intervalo 61-90 tem mais dias ideais. Total de", dias_61_90, "dias")
```
